```python
image_url = 'https://i.pinimg.com/originals/1e/3e/03/1e3e03da41b474a539ebe75220c87b20.jpg'
```

```python
import requests 

image_binary = requests.get(image_url).content

with open('money.jpg', 'wb') as f:
    f.write(image_binary)
```

```python
from PIL import Image

monkey_image = Image.open('money.jpg')
```

```python
monkey_image
```

```python
from pilkit.processors import Thumbnail

processor = Thumbnail(width=300, height=300)
```

```python
thum_image = processor.process(monkey_image)
```

```python
thum_image
```

```python
thum_image.save("thumb-300x300.png")

thum_image.mode = 'RGB'
for quality in [100,80,60,40,20]:
    thum_image.save("thumb-300x300-{}.jpg".format(quality), quality=quality)
```
